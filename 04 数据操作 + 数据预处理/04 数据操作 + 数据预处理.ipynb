```python
import os

import numpy as np
import pandas as pd
import torch
from numpy import nan as NaN
```

```python
os.makedirs(os.path.join('..', 'data'), exist_ok=True)  # data
datafile = os.path.join('..', 'data', 'house_tiny.csv')  # 
```

```python
with open(datafile, 'w') as f:  # 
    f.write('NumRooms,Alley,Price\n')  # 
    f.write('NA,Pave,127500\n')  # 1
    f.write('2,NA,106000\n')  # 2
    f.write('4,NA,178100\n')  # 3
    f.write('NA,NA,140000\n')  # 4

data = pd.read_csv(datafile)  
# NANaN
print('1.:\n', data)
```

```python
inputs, outputs = data.iloc[:, 0: 2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean())  # NaN
print(inputs)
print("-"*40)
print(outputs)
print("-"*40)
# pandasget_dummies
# [ inputs  NaN ] 
# Alley Pave  NaN
inputs = pd.get_dummies(inputs, dummy_na=True)
print('2.pandasget_dummies:\n', inputs)
```

```python
x, y = torch.tensor(inputs.values), torch.tensor(outputs.values)
print('3.')
print(x)
print(y)
```

```python
# fillna
df1 = pd.DataFrame([[1, 2, 3], [NaN, NaN, 2], [NaN, NaN, NaN], [8, 8, NaN]])  # 
print('4.fillna')
print(df1)
print("-"*40)
print(df1.fillna(100))  #  
print("-"*40)
print(df1.fillna({0: 10, 1: 20, 2: 30}))  # 
print("-"*40)
print(df1.fillna(method='ffill'))  # 
# inplace= True
# print(df1.fillna(0, inplace=True))  
```

```python
df2 = pd.DataFrame(np.random.randint(0, 10, (5, 5)))  # 5*5
df2.iloc[1:4, 3] = NaN
df2.iloc[2:4, 4] = NaN  # 
print(df2)
print("-"*40)
print(df2.fillna(method='bfill', limit=2))  # 
print("-"*40)
print(df2.fillna(method="ffill", limit=1, axis=1)) 
print("-"*40)
```
