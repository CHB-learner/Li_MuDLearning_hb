```python
import torch
print('1.张量的创建')
# ones 函数创建一个具有指定形状的新张量，并将所有元素值设置为 1
x = torch.ones(4)
y = torch.arange(12)
print('x:', x)
print('y:', y)
```

```python
print(y.shape)                  # 访问向量的形状
print('y.numel() :', y.numel())  # 返回张量中元素的总个数
```

```python
y = y.reshape(3, 4)  # 改变一个张量的形状而不改变元素数量和元素值
print(y)
```

```python
z = torch.zeros(2, 3, 4)  # 创建一个张量，其中所有元素都设置为0
print('z:', z)
```

```python
w = torch.randn(2, 3, 4)  # 每个元素都从均值为0、标准差为1的标准高斯（正态）分布中随机采样。
print('w:', w)
```

```python
q = torch.tensor([[1, 2, 3], [4, 3, 2], [7, 4, 3]])  # 通过提供包含数值的 Python 列表（或嵌套列表）来为所需张量中的每个元素赋予确定值
print('q:', q)
```

```python
print('2.张量的运算')
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x ** y)  # **运算符是求幂运算
print(torch.exp(x))
```

```python
X = torch.arange(12, dtype=torch.float32).reshape(3, 4)
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print("X",X)
print("Y",Y)
print('cat操作 dim=0', torch.cat((X, Y), dim=0))
print('cat操作 dim=1', torch.cat((X, Y), dim=1))  
# 连结（concatenate） ,将它们端到端堆叠以形成更大的张量。
```

```python
print("X",X)
print("Y",Y)
print('X == Y', X == Y)  # 通过 逻辑运算符 构建二元张量
print('X < Y', X < Y)
print('张量所有元素的和:', X.sum())  # 张量所有元素的和为标量
```

```python
print('3.广播机制')
a = torch.arange(3).reshape(3, 1)
b = torch.arange(2).reshape(1, 2)
print('a:', a)
print('b:', b)
print('a + b:', a + b)  # 神奇的广播运算
```

```python
print('4.索引和切片')
X = torch.arange(12, dtype=torch.float32).reshape(3, 4)
print('X:', X)
print('X[-1]:', X[-1])  # 用 [-1] 选择最后一个元素
print('X[1:3]:', X[1:3])  # 用 [1:3] 选择第二个和第三个元素]
```

```python
X[1, 2] = 9  # 写入元素。
print('X:', X)
X[0:2, :] = 12  # 写入元素。
print('X:', X)
```

```python
print('5.节约内存')
before = id(Y)  # id()函数提供了内存中引用对象的确切地址
Y = Y + X
print(id(Y) == before)
```

```python
before = id(X)
X += Y
print(id(X) == before)  
# 使用 X[:] = X + Y 或 X += Y 来减少操作的内存开销。
```

```python
before = id(X)
X[:] = X + Y
print(id(X) == before)  
# 使用 X[:] = X + Y 或 X += Y 来减少操作的内存开销。
```

```python
print('6.转换为其他 Python对象')
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print("type(Y):",type(Y)) 

A = Y.numpy()
print(type(A))  # 打印A的类型
print(A)

B = torch.tensor(A)
print(type(B))  # 打印B的类型
print(B)
```

# 数据预处理

```python
import os

import numpy as np
import pandas as pd
import torch
from numpy import nan as NaN
```

```python
os.makedirs(os.path.join('..', 'data'), exist_ok=True)  # 在上级目录创建data文件夹
datafile = os.path.join('..', 'data', 'house_tiny.csv')  # 创建文件
```

```python
with open(datafile, 'w') as f:  # 往文件中写数据
    f.write('NumRooms,Alley,Price\n')  # 列名
    f.write('NA,Pave,127500\n')  # 第1行的值
    f.write('2,NA,106000\n')  # 第2行的值
    f.write('4,NA,178100\n')  # 第3行的值
    f.write('NA,NA,140000\n')  # 第4行的值

data = pd.read_csv(datafile)  
# 可以看到原始表格中的空值NA被识别成了NaN
print('1.原始数据:\n', data)
```

```python
inputs, outputs = data.iloc[:, 0: 2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean())  # 用均值填充NaN
print(inputs)
print("-"*40)
print(outputs)
print("-"*40)
# 利用pandas中的get_dummies函数来处理离散值或者类别值。
# [对于 inputs 中的类别值或离散值，我们将 “NaN” 视为一个类别。] 
#由于 “Alley”列只接受两种类型的类别值 “Pave” 和 “NaN”
inputs = pd.get_dummies(inputs, dummy_na=True)
print('2.利用pandas中的get_dummies函数处理:\n', inputs)
```

```python
x, y = torch.tensor(inputs.values), torch.tensor(outputs.values)
print('3.转换为张量：')
print(x)
print(y)
```

```python
# 扩展填充函数fillna的用法
df1 = pd.DataFrame([[1, 2, 3], [NaN, NaN, 2], [NaN, NaN, NaN], [8, 8, NaN]])  # 创建初始数据
print('4.函数fillna的用法：')
print(df1)
print("-"*40)
print(df1.fillna(100))  # 用常数填充 ，默认不会修改原对象
print("-"*40)
print(df1.fillna({0: 10, 1: 20, 2: 30}))  # 通过字典填充不同的常数，默认不会修改原对象
print("-"*40)
print(df1.fillna(method='ffill'))  # 用前面的值来填充
# inplace= True直接修改原对象
# print(df1.fillna(0, inplace=True))  
```

```python
df2 = pd.DataFrame(np.random.randint(0, 10, (5, 5)))  # 随机创建一个5*5
df2.iloc[1:4, 3] = NaN
df2.iloc[2:4, 4] = NaN  # 指定的索引处插入值
print(df2)
print("-"*40)
print(df2.fillna(method='bfill', limit=2))  # 限制填充个数
print("-"*40)
print(df2.fillna(method="ffill", limit=1, axis=1)) 
print("-"*40)
```
